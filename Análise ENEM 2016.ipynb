```python
colunas = ['SG_UF_RESIDENCIA','IN_BAIXA_VISAO','IN_CEGUEIRA','IN_SURDEZ','IN_DEFICIENCIA_AUDITIVA','IN_SURDO_CEGUEIRA','IN_DEFICIENCIA_FISICA','IN_DEFICIENCIA_MENTAL','IN_DEFICIT_ATENCAO','IN_DISLEXIA','IN_DISCALCULIA','IN_AUTISMO','IN_VISAO_MONOCULAR','IN_OUTRA_DEF']

dados = pd.read_csv('../input/microdados_enem_2016_coma.csv', encoding='latin-1', sep=',', usecols=colunas)
```

```python
df = dados
```

```python
total = df.isnull().sum().sort_values(ascending = False)
percentual = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
faltantes  = pd.concat([total, percentual], axis=1, keys=['Total', 'Percentual'])
faltantes
```

```python
df = df.groupby(df.SG_UF_RESIDENCIA).sum()
```

```python
df.reset_index()
```

```python
df.plot(x='SG_UF_RESIDENCIA',y='IN_DISLEXIA', kind='pie', figsize=(8,8)).legend(bbox_to_anchor=(1, 1))
plt.ylabel("")
plt.title("Candidatos portadores de Dislexia / UF - 2016")
plt.show()
```
